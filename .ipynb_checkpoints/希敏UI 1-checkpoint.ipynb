```python
from functools import reduce
import cv2
from PIL import Image, ImageTk, ImageFilter # 导入图像处理函数库
import tkinter as tk
from tkinter import filedialog   #导入文件对话框函数库
import numpy as np

# 创建窗口 设定大小并命名
window = tk.Tk()
window.title('图像显示界面')
window.geometry('600x500')
global img_png           # 定义全局变量 图像的
var = tk.StringVar()    # 这时文字变量储存器
global Img, file_path
global lable_Img

# 创建打开图像和显示图像函数
def Open_Img():
    global img_png, Img, file_path
    OpenFile = tk.Tk() #创建新窗口
    OpenFile.withdraw()
    file_path = filedialog.askopenfilename()
    Img = Image.open(file_path)
    img_png = ImageTk.PhotoImage(Img)
    var.set('已打开\nyeah')
def Show_Img():
    global img_png, Img, label_img
    var.set('已显示')   # 设置标签的文字为 'you hit me'
    img_png = ImageTk.PhotoImage(Img)
    label_Img = tk.Label(window, image=img_png)
    label_Img.configure(image=img_png)
    label_Img.pack()
    
def Forget_Img():
    global img_png, Img, label_img
    var.set('已消除')
    label_Img.pack_forget()
    
# def blur(n, img):
#     global img_png
#     img_png = img_png.filter(ImageFilter.GaussianBlur(5))  # 套用 GaussianBlur，設定模糊半徑為 5
#     img_png.save('i.jpg')
#     img_png = Image.open('i.jpg')
    
def blur():
    global img_png, Img
    Img = Img.filter(ImageFilter.BoxBlur(5))
    img_png = ImageTk.PhotoImage(Img)
#     return out_img

def rotate():
    global img_png, Img
    Img = Img.rotate(45) 
    img_png = ImageTk.PhotoImage(Img)
    
def resize():
    global Img
    Img = Img.resize((128, 128))
    
def transpose():
    global Img
    Img = Img.transpose(Image.Transpose.FLIP_LEFT_RIGHT)

def gray():
    global Img, file_path
    img = cv2.imread(file_path)
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY); # 轉換前，都先將圖片轉換成灰階色彩
    ret, output1 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)     # 如果大於 127 就等於 255，反之等於 0。
    Img = Image.fromarray(cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB))

def normalize():
    global Img, file_path
    img2 = cv2.imread(file_path)
    final_img = cv2.normalize(img2,  None, 0, 255, cv2.NORM_MINMAX)
    Img = Image.fromarray(cv2.cvtColor(final_img, cv2.COLOR_BGR2RGB))


# def resize


# 创建文本窗口，显示当前操作状态
Label_Show = tk.Label(window,
    textvariable=var,   # 使用 textvariable 替换 text, 因为这个可以变化
    bg='yellow', font=('Arial', 12), width=15, height=2)
Label_Show.pack()
# 创建打开图像按钮
btn_Open = tk.Button(window,
    text='Open',      # 显示在按钮上的文字
    width=15, height=2,
    command=Open_Img)     # 点击按钮式执行的命令
btn_Open.pack()    # 按钮位置
# 创建显示图像按钮
btn_Show = tk.Button(window,
    text='Show',      # 显示在按钮上的文字
    width=15, height=2,
    command=Show_Img)     # 点击按钮式执行的命令
btn_Show.pack()    # 按钮位置

#模糊圖片
btn_blur = tk.Button(window,
    text='Blur',      # 显示在按钮上的文字
    width=15, height=2,
    command=blur)     # 点击按钮式执行的命令
btn_blur.pack()    # 按钮位置

#旋轉
btn_rotate = tk.Button(window,
    text='Rotate',      # 显示在按钮上的文字
    width=15, height=2,
    command=rotate)     # 点击按钮式执行的命令
btn_rotate.pack()    # 按钮位置

#resize
btn_resize = tk.Button(window,
    text='Resize',      # 显示在按钮上的文字
    width=15, height=2,
    command=resize)     # 点击按钮式执行的命令
btn_resize.pack()    # 按钮位置

#transpose
btn_transpose = tk.Button(window,
    text='左右轉置',      # 显示在按钮上的文字
    width=15, height=2,
    command=transpose)     # 点击按钮式执行的命令
btn_transpose.pack()    # 按钮位置

#灰了
btn_gray = tk.Button(window,
    text='Gray',      # 显示在按钮上的文字
    width=15, height=2,
    command=gray)     # 点击按钮式执行的命令
btn_gray.pack()    # 按钮位置

#normalize
btn_normalize = tk.Button(window,
    text='Normalize',      # 显示在按钮上的文字
    width=15, height=2,
    command=normalize)     # 点击按钮式执行的命令
btn_normalize.pack()    # 按钮位置

btn_Forget = tk.Button(window,
                       text = 'Click to hide Label',
                       width=15, height=2,
                       command = Forget_Img)
btn_Forget.pack()

# 运行整体窗口
window.mainloop()
```
